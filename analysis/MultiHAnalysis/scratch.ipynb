```python
import boost_histogram as bh
```

```python
def get_boost(n, bins, v=None):
    if v is None: v = np.zeros_like(n)
    H = bh.Histogram(bh.axis.Variable(bins), storage=bh.storage.Weight())
    H[...] = np.stack([n, v]).T
    return H
```

```python
from matplotlib.gridspec import GridSpec
```

```python
bins = np.arange(100)
x = (bins[:-1]  + bins[1:])/2
boost_dict = {}

fig = plt.figure()
gs = GridSpec(2, 1, height_ratios=[4,1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

n_ttbar = Hist(f['Events']['Pileup_nTrueInt'].array(), bins=np.arange(100), ax=ax1, density=True)
n_nom = Hist(x, weights=nom, bins=np.arange(100), ax=ax1, density=True)
n_up = Hist(x, weights=up, bins=np.arange(100), ax=ax1, density=True)
n_dn = Hist(x, weights=dn, bins=np.arange(100), ax=ax1, density=True)

r_nom = n_nom/n_ttbar
r_nom = np.where(n_nom == n_ttbar, 1, r_nom)
r_nom = np.where(np.isinf(r_nom), 1, r_nom)
boost_dict['PUweights'] = get_boost(r_nom, bins, v=np.zeros_like(r_nom))

r_up = n_up/n_ttbar
r_up = np.where(n_up == n_ttbar, 1, r_up)
r_up = np.where(np.isinf(r_up), 1, r_up)
boost_dict['PUweights_up'] = get_boost(r_up, bins, v=np.zeros_like(r_up))

r_dn = n_dn/n_ttbar
r_dn = np.where(n_dn == n_ttbar, 1, r_dn)
r_dn = np.where(np.isinf(r_dn), 1, r_dn)
boost_dict['PUweights_down'] = get_boost(r_dn, bins, v=np.zeros_like(r_dn))

r_nom = Hist(x, weights=r_nom, bins=bins, ax=ax2)
```
